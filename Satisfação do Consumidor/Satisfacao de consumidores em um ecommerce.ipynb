UNIÃO DOS DATASETS

Merge de todos os datasets referentes ao **consumidor**
(olist_orders_dataset.csv, olist_order_reviews_dataset.csv, olist_order_payments_dataset.csv)


```python
consumidor = pd.concat([orders, avaliacoes, pagamentos],ignore_index=True)
```

```python
consumidor.head(3)
```

```python
consumidor.drop_duplicates(subset="order_id")
```

Merge de todos os datasets referentes ao **vendedor** 

```python
itens_pedidos.head(3)
```

```python
produtos.head(3)
```

```python
vendedores.head(3)
```

```python
#join de items pedidos e produtos por id do produto 
produtos_pedidos = pd.merge(itens_pedidos, produtos, on=["product_id"],how="inner")
```

```python
#join dos produtos pedidos com a tabela de vendedores por seller_id
```

```python
vendedor_produto = pd.merge(produtos_pedidos, vendedores, on=["seller_id"],how="inner")
```

```python
vendedor_produto.head(5)
```

```python
#nome das colunas do dataframe apos o join  
list(vendedor_produto.columns.values)
```

Merge dos datasets do **consumidor e do vendedor**

```python
df_ecommerce = pd.merge(consumidor,vendedor_produto, on=["order_id"],how="inner")
```

```python
df_ecommerce.head(5)
```

```python
list(df_ecommerce.columns.values)
```
